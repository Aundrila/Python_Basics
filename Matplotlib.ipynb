# Matplotlib Exercises

```python
# import numpy library
import numpy as np

# create vector x using arrange  between 0 and 100 with stepsize =3
x = np.arange(0,99,3)


# y=x^2
y= x**2
# z= e^x
z= np.exp(x)
# h=sin(x)
h= np.sin(x)
# g=x^2
g= x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

##### Exercise 1

```python

# create a figure object  and call it fig
fig = plt.figure()
# add an axis to the figure canvas at [0.2,0.2,0.8,0.8].


#Plot x, y,g on that axes and set the labels ,titles  legend. set different color and marker for y and g
axes = fig.add_axes([0.2, 0.2, 0.8, 0.8])
axes.plot(x , y , "b", marker = "o")
axes.plot(x , g , "b", marker = "o")
```

###### Exercise 2

```python
#create the plots below
# the size of the figure is (12,7)
fig, ax = plt.subplots(2, 2, figsize = (12, 7))
ax[0][0].scatter(x, y, color = "r", marker = "*")
ax[0][1].plot(x, z, color = "b", marker = "+")
ax[1][0].plot(x, h, color = "g")
ax[1][1].scatter(x, g, c = g)
```
